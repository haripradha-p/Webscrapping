# Python program to display all the header tags from Wikipedia.org

```python
import requests
from bs4 import BeautifulSoup


response = requests.get("https://www.wikipedia.org")


soup = BeautifulSoup(response.content, "html.parser")

header_tags = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])

# Print the text content of each header tag
for tag in header_tags:
    print(tag.text.strip())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


response = requests.get("https://www.imdb.com/chart/top")


soup = BeautifulSoup(response.content, "html.parser")


movie_tags = soup.find_all("td", class_="titleColumn")

# Create lists to store the movie data
names = []
ratings = []
years = []

# Loop through each movie tag and extract its data
for tag in movie_tags:
    # Extract the movie name
    name = tag.a.text
    names.append(name)
    
    # Extract the movie year of release
    year = tag.span.text.strip("()")
    years.append(year)
    
    # Extract the movie rating
    rating = tag.parent.find("td", class_="ratingColumn imdbRating").strong.text
    ratings.append(rating)

# Create a DataFrame from the movie data
df = pd.DataFrame({"Name": names, "Year": years, "Rating": ratings})

# Print the DataFrame
print(df)
```

# A python program to scrape cricket rankings from icc-cricket.com.  to scrape:
a)	Top 10  ODI teams in men’s cricket along with the records for matches, points and rating
b)	Top 10 ODI Batsmen along with the records of their team and rating.
c)	Top 10 ODI bowlers along with the records of their team and rating


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Make a GET request to the ODI team rankings page
team_response = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")

# Parse the HTML content using BeautifulSoup
team_soup = BeautifulSoup(team_response.content, "html.parser")

# Find the table containing the top 10 ODI teams
team_table = team_soup.find("table", class_="table")

# Create lists to store the team data
team_names = []
team_matches = []
team_points = []
team_ratings = []

# Loop through each row in the table and extract the team data
for row in team_table.find_all("tr")[1:11]:
    # Extract the team name from the second cell in the row
    name = row.find_all("td")[1].text.strip()
    team_names.append(name)
    
    # Extract the number of matches played from the third cell in the row
    matches = row.find_all("td")[2].text.strip()
    team_matches.append(matches)
    
    # Extract the number of points from the fourth cell in the row
    points = row.find_all("td")[3].text.strip()
    team_points.append(points)
    
    # Extract the team rating from the fifth cell in the row
    rating = row.find_all("td")[4].text.strip()
    team_ratings.append(rating)

# Create a DataFrame from the team data
team_df = pd.DataFrame({
    "Team": team_names,
    "Matches": team_matches,
    "Points": team_points,
    "Rating": team_ratings
})

# Print the team DataFrame
print("Top 10 ODI Teams:\n")
print(team_df)

# Make a GET request to the ODI batsmen rankings page
batsmen_response = requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")

# Parse the HTML content using BeautifulSoup
batsmen_soup = BeautifulSoup(batsmen_response.content, "html.parser")

# Find the table containing the top 10 ODI batsmen
batsmen_table = batsmen_soup.find("table", class_="table")

# Create lists to store the batsmen data
batsmen_names = []
batsmen_teams = []
batsmen_ratings = []

# Loop through each row in the table and extract the batsmen data
for row in batsmen_table.find_all("tr")[1:11]:
    # Extract the batsman name from the second cell in the row
    name = row.find_all("td")[1].text.strip()
    batsmen_names.append(name)
    
    # Extract the batsman team from the third cell in the row
    team = row.find_all("td")[2].text.strip()
    batsmen_teams.append(team)
    
    # Extract the batsman rating from the fifth cell in the row
    rating = row.find_all("td")[4].text.strip()
    batsmen_ratings.append(rating)

# Create a DataFrame from the batsmen data
batsmen_df = pd.DataFrame({
    "Batsman": batsmen_names,
    "Team": batsmen_teams,
    "Rating": batsmen_ratings
})

# Print the batsmen DataFrame
print("\nTop 10 ODI Batsmen:\n")
print(batsmen_df)

# Make
```

```python
import requests
from bs4 import BeautifulSoup

# Make a GET request to the ICC cricket rankings page for ODI bowlers
response = requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling")

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the top 10 ODI bowlers data
table = soup.find("table", class_="table rankings-table")

# Find all the rows in the table except the first one (header row)
rows = table.find_all("tr")[1:11]

# Loop through each row and extract the player name, team, and rating
for row in rows:
    # Extract the player name from the second cell in the row
    name = row.find_all("td")[1].find("a").text.strip()

    # Extract the team name from the third cell in the row
    team = row.find_all("td")[2].text.strip()

    # Extract the rating from the fifth cell in the row
    rating = row.find_all("td")[4].text.strip()

    # Print the player name, team, and rating
    print(name + " - " + team + " - " + rating)
```

# A python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world:
i)Headline
ii)Time
iii) News Link


```python
import requests
from bs4 import BeautifulSoup

url = "https://www.cnbc.com/world/?region=world"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

articles = soup.find_all("div", {"class": "LatestNews-container"})

for article in articles:
    headline = article.find("a", {"class": "LatestNews-headline"}).text.strip()
    time = article.find("time", {"class": "LatestNews-timestamp"}).text.strip()
    news_link = article.find("a", {"class": "LatestNews-headline"})["href"]
    print(f"Headline: {headline}\nTime: {time}\nNews Link: {news_link}\n")

```

# A python program to scrape the details of most downloaded articles from AI in last 90 days https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles

ii)	Authors
iii)	Published Date
iv)	Paper URL


```python
import requests
from bs4 import BeautifulSoup

url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")

articles = soup.find_all("li", {"class": "sc-9zxyh7-1 sc-9zxyh7-2 kOEIEO hvoVxs"})

for article in articles:
    title = article.find("h2").text.strip()
    authors = article.find("span").text.strip()
    date = article.find("span").text.strip()
    paper_url = article.find("a")["href"]
    print(f"Title: {title}\nAuthors: {authors}\nDate: {date}\nPaper URL: {paper_url}\n")


```

# 7.	A python program to scrape mentioned details from dineout.co.in
i)	Restaurant name
ii)	Cuisine
iii)	Location
iv)	Ratings
v)	ImageURL


```python
import requests
from bs4 import BeautifulSoup

# Make a GET request to the dineout.co.in page
response = requests.get("https://www.dineout.co.in/bangalore-restaurants")

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the list of restaurants
restaurant_list = soup.find("div", class_="searchResultContainer")

# Check if the restaurant list was found
if restaurant_list:
    # Find all the restaurants in the list
    restaurants = restaurant_list.find_all("div", class_="restntInfo")

    # Loop through each restaurant and extract the details
    for restaurant in restaurants:
        # Extract the name of the restaurant
        name = restaurant.find("a", class_="restntName").text.strip()

        # Extract the cuisine of the restaurant
        cuisine = restaurant.find("div", class_="restntCuisine").text.strip()

        # Extract the location of the restaurant
        location = restaurant.find("div", class_="restntLoc").text.strip()

        # Extract the ratings of the restaurant
        ratings = restaurant.find("span", class_="ratCnt").text.strip()

        # Extract the image URL of the restaurant
        image_url = restaurant.find("img")["data-src"]

        # Print the details
        print("Name:", name)
        print("Cuisine:", cuisine)
        print("Location:", location)
        print("Ratings:", ratings)
        print("Image URL:", image_url)
        print()
else:
    print("Restaurant list not found.")
```

```python
import requests
from bs4 import BeautifulSoup

# Make a GET request to the Google Scholar top publications page
response = requests.get("https://scholar.google.com/citations?view_op=top_venues&h1=en")

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the table of top publications
table = soup.find("table")
print(table)

# Check if the table was found
if table:
    # Find all the rows in the table except the first one (header row)
    rows = table.find_all("tr")[1:]

    # Loop through each row and extract the rank, publication name, h5-index, and h5-median
    for row in rows:
        # Extract the rank from the first cell in the row
        rank = row.find("td", class_="gs_rnk_cell").text.strip()

        # Extract the publication name from the second cell in the row
        publication = row.find("a", class_="gs_rnk_link").text.strip()

        # Extract the h5-index from the third cell in the row
        h5_index = row.find_all("td")[2].text.strip()

        # Extract the h5-median from the fourth cell in the row
        h5_median = row.find_all("td")[3].text.strip()

        # Print the rank, publication name, h5-index, and h5-median
        print(rank)
        print(publication)
        print(h5_index)
        print(h5_median)
        print()
else:
    print("Table not found.")
```
